### example exercise to plotting in Matplotlib

#### Ohio State’s leading causes of death from the year 2012.
You have been provided with a dataset that lists Ohio State’s leading causes of death from the year 2012. Using the two data points:

Cause of deaths and
Percentile
Draw a pie chart to visualize the dataset

```python
# the data
data={'Cause of deaths':['Chronic Disease','Unintetional Injuries','Alzheimers','Infuenza and Pneumonia','Sepsis','Others'],
     'Percentile':[62, 5, 4, 2, 1, 26]}

df=pd.DataFrame(data) # creating dataframe
df
```

```python
df.sort_values(by=['Percentile'],inplace=True,ascending=False)
df
```

```python
plt.figure(figsize=(7,7))
plt.title= 'Ohio State’s leading causes of death from the year 2012'
plt.pie(df['Percentile'], labels=df['Cause of deaths'], autopct='%1.0f%%')
```

```python
df2 = pd.read_csv('house-prices.csv')
```

```python
plt.figure(figsize=(15,10),dpi=60)
plt.title ="Scatter Plot for Price and Sqft"
plt.xlabel="Price"
plt.ylabel="SqFt"
plt.scatter(x=df2['Price'], y=df2['SqFt'])
plt.show()
# plt.savefig("Scatter Plot for Price over Sqft.png")
```

# Plotting 3D

```python
plt.figure(figsize=(15,10),dpi=50)
axes= plt.axes(projection='3d')
axes.scatter(df2['Home'],df2['Price'],df2['SqFt'])
axes.set_xlabel="Home"
axes.set_ylable="Price"
axes.set_zlabel="SqFt"
plt.show()
```
