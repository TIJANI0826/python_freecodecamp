```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dfExcel = pd.read_excel('../samplefiles/Mortgages+Data+Model.xlsx')
```

```python
x_values = list(range(1000))
squares = [x**2 for x in x_values]
plt.scatter(x_values, squares, s=0.5)
plt.show()
```

```python
# scaling using s, fontsize using fontsize setting x and y axis labels using xlabel and y
# ylabel respectively
x_values = list(range(1000))
squares = [x**2 for x in x_values]
plt.scatter(x_values, squares, s=10)
plt.title("Square Numbers", fontsize=24)
plt.xlabel("Value", fontsize=18)
plt.ylabel("Square of Value", fontsize=18)
plt.tick_params(axis='both', which='major',
labelsize=14)
plt.axis([0, 1100, 0, 1100000])
plt.show()
```

```python
# Colormap  using parameter c
plt.scatter(x_values, squares, c=squares,
 cmap=plt.cm.Blues, edgecolor='none',
 s=10)
```
